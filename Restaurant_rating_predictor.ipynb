```python
y=df["Aggregate rating"]
```

```python
x
```

```python
x["Price range"].value_counts
```

```python
from sklearn import preprocessing
labelencoder=preprocessing.LabelEncoder()
```

```python
x["Has Table booking"]=labelencoder.fit_transform(x["Has Table booking"])
```

```python
x
```

```python
x["Has Online delivery"]=labelencoder.fit_transform(x["Has Online delivery"])
```

```python
x
```

```python
  from sklearn.preprocessing import StandardScaler
  scaler=StandardScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
```

```python
def modelresults(predictions):
  print("Mean absolute error is {}".format(mean_absolute_error(y_test,predictions)))
  print("Root mean squared error is {}".format(np.sqrt(mean_squared_error(y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predictionsfroml=lr.predict(x_test)
```

```python
modelresults(predictionsfroml)
```

```python
predictionsfroml
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel=SVR()
```

```python
param_gridsvr={"C":[0.1,0.5],"kernel":["linear","poly"],"degree":[2]}
```

```python
gridsvr=GridSearchCV(svrmodel,param_gridsvr)
gridsvr.fit(x_train,y_train)
```

```python
predictionsvr=gridsvr.predict(x_test)
```

```python
modelresults(predictionsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid={
    "max_depth":[10,30],
    "min_samples_leaf":[1,2],
    "min_samples_split":[2,5]
}
```

```python
treemodel=DecisionTreeRegressor()
```

```python
grid_tree=GridSearchCV(estimator=treemodel,param_grid=param_grid)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
treepredictions=grid_tree.predict(x_test)
```

```python
modelresults(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel=RandomForestRegressor()
```

```python
param_gridrfr={
    "max_depth":[5,15],"n_estimators":[2,5,10]}
```

```python
gridrfr=GridSearchCV(rfrmodel,param_gridrfr)
```

```python
gridrfr.fit(x_train,y_train)
```

```python
randomforestpredictions=gridrfr.predict(x_test)
```

```python
modelresults(randomforestpredictions)
```

```python
gridrfr.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn_param_grid={"n_neighbors":[9,11,13,15,17]}
knn_grid_search= GridSearchCV(knn,knn_param_grid)
```

```python
knn_grid_search.fit(x_train,y_train)
```

```python
knnprediction=knn_grid_search.predict(x_test)
```

```python
modelresults(knnprediction)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
ada_param_grid={"n_estimators":[50,100,200],
                "learning_rate":[0.1,0.5,1]}
ada_grid_search=GridSearchCV(ada,ada_param_grid)
```

```python
ada_grid_search.fit(x_train,y_train)
```

```python
adaprediction=ada_grid_search.predict(x_test)
```

```python
modelresults(adaprediction)
```
